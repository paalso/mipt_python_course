Постройте график функции

y(x) = x*x - x - 6

и по графику найдите найдите корни уравнения y(x) = 0. (Не нужно применять численных методов — просто приблизьте график к корням функции настолько, чтобы было удобно их найти.)

```python
import numpy as np
import matplotlib.pyplot as plt


def y(x):
    return x*x - x - 6

min_, max_ = -7, 7
min_, max_ = -2.2, 3.1
x = np.arange(min_, max_, 0.01) # генерирует вектор аргументов
plt.figure(figsize=(9, 6))

plt.plot(x, y(x), label=r'$y(x)=x*x - x - 6$')
plt.hlines(0, min_, max_, 'r')  # горизонтальная линия y = 0

# подписи осей
plt.xlabel(r'$x$', fontsize=14)
plt.ylabel(r'$y(x)$', fontsize=14)

# Заголовок
plt.title(r'График функции $y(x)=x*x - x - 6$')
plt.grid(True) # рисовать решетку

plt.show()
```

Постройте график функции

from IPython.display import Image
![](http://cs.mipt.ru/python/images/lab1/task3.png)

```python
import numpy as np
import matplotlib.pyplot as plt
from math import tan, sin, exp, log


def f(x):
    return log((x ** 2 + 1) * exp(- abs(x) / 10), 1+tan(1 / (1 + sin(x) ** 2)))

lim = 100
min_, max_ = -lim, lim
x = np.arange(min_, max_, 0.01) # генерирует вектор аргументов
y = np.vectorize(f)
plt.figure(figsize=(20, 6))

plt.plot(x, y(x))

# подписи осей
plt.xlabel(r'$x$', fontsize=14)
plt.ylabel(r'$y(x)$', fontsize=14)

# Заголовок
plt.grid(True) # рисовать решетку

plt.show()
```

Используя функцию eval() постройте график функции, введённой с клавиатуры. Чтобы считать данные с клавиатуры, используйте функцию input(). Попробуйте включить эффект «рисование от руки» посредством вызова plt.xkcd()

```python
import numpy as np
import matplotlib.pyplot as plt
from math import *

func_str = input("Функция (от аргумента x): ")
min_, max_ = map(int, input("Диапазон (min, max): ").split())


def f(x):
    return eval(func_str)


x = np.arange(min_, max_, 0.01) # генерирует вектор аргументов
y = np.vectorize(f)


with plt.xkcd():
    plt.figure(figsize=(20, 6))
    plt.plot(x, y(x), label=func_str)

    # подписи осей
    plt.xlabel(r'$x$', fontsize=14)
    plt.ylabel(r'$y(x)$', fontsize=14)

    # Заголовок
    plt.title('График функции y(x) = {}'.format(func_str))
    plt.grid(True) # рисовать решетку - не работает в режиме 'рисмования от руки'

    plt.show(func_str)
```
