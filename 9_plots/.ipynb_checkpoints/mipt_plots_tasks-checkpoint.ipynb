Найдите значения в точках 1, 10, 1000.

```python
# math version

import math


def y(x):
    return math.log(math.exp(1 / (math.sin(x) + 1)) / (1.25 + 1 / x ** 15), 1 + x * x)


print(y(1))
print(y(10))
print(y(1000))
```

```python
# numpy version

import numpy as np


def y(x):
    return np.log(np.exp(1 / (np.sin(x) + 1)) / (1.25 + 1 / x ** 15)) / np.log(1 + x * x)


print(y(1))
print(y(10))
print(y(1000))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def y(x):
    return np.log(np.exp(1 / (np.sin(x) + 1)) / (1.25 + 1 / x ** 15)) / np.log(1 + x * x)

x = np.arange(1, 1100, 0.01) # генерирует вектор аргументов
plt.figure(figsize=(9, 6))

# легенда
plt.plot(x, y(x))
# plt.legend(loc='best', fontsize=12)

# подписи осей
plt.xlabel(r'$x$', fontsize=14)
plt.ylabel(r'$y(x)$', fontsize=14)

# Заголовок
plt.title(r'Графики функций $f_1(x)=\sin(x),\ f_2(x)=\cos(x),\ f_3(x)=-x$')
plt.grid(True) # рисовать решетку

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def y(x):
   return np.log(np.exp(1 / (np.sin(x) + 1)) / (1.25 + 1 / x ** 15)) / np.log(1 + x * x)

x = np.arange(0, 2, 0.01)
t = np.arange(8, 12, 0.01)
u = np.arange(900, 1100, 0.01)

#subplot 1


#subplot 2
sp = plt.subplot(222)
plt.plot(x, y(x))
plt.title('x = 0..2')
plt.vlines(1, -600000, 10000, color = 'r')
plt.grid(True)

#subplot 3
sp = plt.subplot(223)
plt.plot(t, y(t))
plt.title('x = 8..12')
plt.vlines(10, -10, 150, color = 'r')
plt.grid(True)

#subplot 4
sp = plt.subplot(224)
plt.plot(u, y(u))
plt.title('x = 1000..1200')
plt.vlines(1000, -1, 50, color = 'r')
plt.grid(True)

plt.show()
```

### Упражнение 2

остройте график функции

y(x) = x*x - x - 6

и по графику найдите найдите корни уравнения y(x) = 0. (Не нужно применять численных методов — просто приблизьте график к корням функции настолько, чтобы было удобно их найти.)

```python
import numpy as np
import matplotlib.pyplot as plt


def y(x):
    return x*x - x - 6

min_, max_ = -7, 7
min_, max_ = -2.2, 3.1
x = np.arange(min_, max_, 0.01) # генерирует вектор аргументов
plt.figure(figsize=(9, 6))

plt.plot(x, y(x), label=r'$y(x)=x*x - x - 6$')
plt.hlines(0, min_, max_, 'r')  # горизонтальная линия y = 0

# подписи осей
plt.xlabel(r'$x$', fontsize=14)
plt.ylabel(r'$y(x)$', fontsize=14)

# Заголовок
plt.title(r'График функции $y(x)=x*x - x - 6$')
plt.grid(True) # рисовать решетку

plt.show()
```
